```python
# Count total rides by month for each user type

monthly_count = df.pivot_table(
    index="month_name",
    columns="member_casual",
    values="ride_id",
    aggfunc="count"
)

# Arrange months in calendar order

month_order = [
    "June","July","August","September",
    "October","November","December",
    "January","February","March","April","May"
]

monthly_count = monthly_count.reindex(month_order)

monthly_count
```

```python
# Calculate percentage distribution within each user type

monthly_percent = (
    monthly_count /
    monthly_count.sum()
    * 100
).round(2)

monthly_percent
```

```python
# Count total rides for each hour by user type

hourly_count = df.pivot_table(
    index="hour",
    columns="member_casual",
    values="ride_id",
    aggfunc="count"
)

# Display the table

hourly_count
```

```python
# Find the top 10 most frequently used start stations

top_start_overall = (
    df["start_station_name"]
    .value_counts()
    .head(10)
)

top_start_overall
```

```python
# Top 10 start stations used by annual members

member_start = (
    df[df["member_casual"] == "member"]["start_station_name"]
    .value_counts()
    .head(10)
)

member_start
# Top 10 start stations used by casual riders

casual_start = (
    df[df["member_casual"] == "casual"]["start_station_name"]
    .value_counts()
    .head(10)
)

casual_start
# Compare the top start stations of members and casual riders

top_start_compare = pd.concat(
    [member_start, casual_start],
    axis=1,
    keys=["Member", "Casual"]
)

top_start_compare
```

```python
# Reset index so station names become a normal column
member_df = member_start.reset_index()
member_df.columns = ["Member Start Station", "Member Rides"]

casual_df = casual_start.reset_index()
casual_df.columns = ["Casual Start Station", "Casual Rides"]

# Merge side by side
top_start_report = pd.concat([member_df, casual_df], axis=1)

top_start_report.index = range(1, len(top_start_report) + 1)
top_start_report.index.name = "Rank"

top_start_report
```

```python
# Find the top 10 end stations used by annual members

member_end = (
    df[df["member_casual"] == "member"]["end_station_name"]
    .value_counts()
    .head(10)
)

member_end
```

```python
# Find the top 10 end stations used by casual riders

casual_end = (
    df[df["member_casual"] == "casual"]["end_station_name"]
    .value_counts()
    .head(10)
)

casual_end
```

```python
# Convert Series into DataFrames
member_end_df = member_end.reset_index()
member_end_df.columns = ["Member End Station", "Member Rides"]

casual_end_df = casual_end.reset_index()
casual_end_df.columns = ["Casual End Station", "Casual Rides"]

# Merge side by side for reporting
top_end_report = pd.concat(
    [member_end_df, casual_end_df],
    axis=1
)

# Add ranking
top_end_report.index = range(1, len(top_end_report) + 1)
top_end_report.index.name = "Rank"

top_end_report
```
